```python

!pip install reportlab pandas

import pandas as pd
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, PageBreak
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet
from google.colab import files

df = pd.read_csv("students.csv")

numeric_cols = df.select_dtypes(include='number')

analysis_text = []

if not numeric_cols.empty:
    for col in numeric_cols.columns:
        avg = df[col].mean()
        mx = df[col].max()
        mn = df[col].min()
        analysis_text.append(f"{col} → Avg: {round(avg,2)}, Max: {mx}, Min: {mn}")
else:
    analysis_text.append("No numeric data found for analysis.")

doc = SimpleDocTemplate("final_report.pdf")
styles = getSampleStyleSheet()
elements = []

elements.append(Paragraph("CSV Data Analysis Report", styles['Title']))
elements.append(Spacer(1, 12))

elements.append(Paragraph("Analysis Summary:", styles['Heading2']))
elements.append(Spacer(1, 8))

for line in analysis_text:
    elements.append(Paragraph(line, styles['Normal']))

elements.append(Spacer(1, 12))

chunk_size = 25

for i in range(0, len(df), chunk_size):
    chunk = df.iloc[i:i+chunk_size]
    table_data = [list(chunk.columns)] + chunk.values.tolist()

    table = Table(table_data)

    table.setStyle(TableStyle([
        ('BACKGROUND', (0,0), (-1,0), colors.grey),
        ('TEXTCOLOR', (0,0), (-1,0), colors.white),
        ('GRID', (0,0), (-1,-1), 1, colors.black),
    ]))

    elements.append(table)
    elements.append(PageBreak())

# Build PDF
doc.build(elements)

print("✅ PDF Generated Successfully!")

# Step 8: Download PDF
files.download("final_report.pdf")
```
