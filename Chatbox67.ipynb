```python
import nltk
import random
import string
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

lemmatizer = WordNetLemmatizer()

intents = {
    "greeting": {
        "patterns": ["hello", "hi", "hey", "good morning", "good evening"],
        "responses": ["Hello! How can I help you?", "Hi there!", "Hey! Ask me anything."]
    },
    "goodbye": {
        "patterns": ["bye", "see you", "goodbye"],
        "responses": ["Goodbye!", "See you later!", "Take care!"]
    },
    "thanks": {
        "patterns": ["thanks", "thank you", "thx"],
        "responses": ["You're welcome!", "No problem!", "Happy to help!"]
    },
    "about": {
        "patterns": ["what is your name", "who are you", "about you"],
        "responses": ["I am a chatbot built using NLTK.", "I am your AI assistant."]
    },
    "weather": {
        "patterns": ["weather", "temperature", "how is weather"],
        "responses": ["I can't fetch real-time weather, but it's always nice to chat!"]
    }
}

def preprocess(sentence):
    tokens = word_tokenize(sentence.lower())
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in string.punctuation]
    return tokens

def match_intent(user_input):
    user_tokens = preprocess(user_input)
    best_match_score = 0
    best_intent_key = None

    for intent_key, intent_data in intents.items():
        for pattern in intent_data["patterns"]:
            pattern_tokens = preprocess(pattern)

            common_words = len(set(user_tokens) & set(pattern_tokens))

            if common_words > best_match_score:
                best_match_score = common_words
                best_intent_key = intent_key

            elif common_words == best_match_score and common_words > 0 and user_tokens == pattern_tokens:
                best_intent_key = intent_key
                return random.choice(intents[best_intent_key]["responses"])

    if best_intent_key and best_match_score > 0:
        return random.choice(intents[best_intent_key]["responses"])
    else:
        return "Sorry, I didn't understand that. Can you rephrase?"


def chatbot():
    print("🤖 Chatbot is running! Type 'quit' to exit.\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "quit":
            print("Bot: Goodbye!")
            break

        response = match_intent(user_input)
        print("Bot:", response)

chatbot()
```
